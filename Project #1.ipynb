```python
plt.figure(figsize=(8,6))
df.plot.scatter(x='Wagers',
                y='Online Sports Wagering Win/(Loss)',
                ax=plt.gca())
plt.title('Win/(Loss) vs. Wagers')
plt.tight_layout()
```

```python
sns.barplot(x='Licensee', y='Patron Winnings', data=df)
plt.title('Patron Winnings by Licensee')
plt.xticks(rotation=45)
```
